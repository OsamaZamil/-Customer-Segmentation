# RFM Analysis:

```python
import datetime as dt

latest_date = df['InvoiceDate'].max() + dt.timedelta(days = 1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,
    'InvoiceNo': 'count',
    'Total': 'sum'
}).reset_index()

rfm.rename(columns = {
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'Total': 'MonetaryValue'
}, inplace = True)
```

```python
rfm["Recency_score"] = pd.qcut(rfm['Recency'], 5, labels = [5, 4, 3, 2, 1])

rfm["Frequency_score"] = pd.qcut(rfm['Frequency'].rank(method = "first"), 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm["rfm_segment"] = rfm['Recency_score'].astype(str) + rfm['Frequency_score'].astype(str)
rfm
```

```python
sns.histplot(data = rfm, x = 'Recency', bins = 30)
plt.title('Recency Distribution')
plt.show()
```

```python
sns.histplot(data = rfm, x = 'Frequency', bins = 30)
plt.title('Frequency Distribution')
plt.show()
```

```python
sns.histplot(data = rfm, x = 'MonetaryValue', bins = 30)
plt.title('Monetary Value Distribution')
plt.show()
```

```python
segment_map = {
    r'[1-2][1-2]' : 'Hibernating',
    r'[1-2][3-4]' : 'At-Risk',
    r'[1-2]5' : "Can\'t Loose",
    r'3[1-2]' : 'About to Slip',
    r'33' : 'Need Attention',
    r'[3-4][4-5]' : 'Loyal Customers',
    r'41' : 'Promising',
    r'51' : 'New Customers',
    r'[4-5][2-3]' : 'Potential Loyalists',
    r'5[4-5]' : 'Champions',
}
```

```python
rfm['rfm_segment'] = rfm['rfm_segment'].replace(segment_map, regex = True)
rfm
```

```python
sns.pairplot(rfm,hue="rfm_segment")
```

# Data Standardization and Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'MonetaryValue']])

rfm_scaled.shape
```

```python
from sklearn.decomposition import PCA
dim_red = PCA()
```

```python
rfm_pca = dim_red.fit_transform(rfm_scaled)
rfm_pca.shape
```

```python
dim_red.components_
```

```python
dim_red.explained_variance_ratio_
```

```python
dim_red.get_feature_names_out()
```

```python
df_pca = pd.DataFrame(rfm_pca,columns=dim_red.get_feature_names_out())
df_pca
```

```python
df_pca["rfm_segment"] = rfm['rfm_segment']
df_pca
```

```python
sns.pairplot(df_pca,hue="rfm_segment")
```

```python
df_pca = df_pca.drop("rfm_segment", axis='columns')
```

# Determining Optimal Number of Clusters:

```python
from sklearn.cluster import KMeans
lo_inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_pca)
    lo_inertia.append(kmeans.inertia_)
    print(i,kmeans.inertia_)
```

```python
plt.plot(range(1, 11),lo_inertia)
```

# K-means Clustering:

```python
kmeans = KMeans(n_clusters=5)
rfm['Cluster'] = kmeans.fit_predict(df_pca)
df_pca['Cluster'] = rfm['Cluster']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_pca['pca1'], df_pca['pca2'], c=df_pca['Cluster'], cmap='viridis')
plt.title('K-means Clustering with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
cluster_analysis = rfm.groupby('Cluster').mean()
print(cluster_analysis)
```

```python
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': ['mean', 'median', 'min', 'max'],
    'Frequency': ['mean', 'median', 'min', 'max'],
    'MonetaryValue': ['mean', 'median', 'min', 'max'],
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'CustomerCount'}).reset_index()

cluster_summary
```

# Cluster Profiling:

```python
cluster_profile = rfm.groupby('Cluster').agg({
    'MonetaryValue': 'mean',
    'Frequency': 'mean',
    'CustomerID': 'count'
}).rename(columns={'MonetaryValue': 'Avg_Spending', 'Frequency': 'Avg_Frequency', 'CustomerID': 'CustomerCount'}).reset_index()

print(cluster_profile)
```

# Evaluation using silhouette score:

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette score
silhouette_avg = silhouette_score(df_pca.iloc[:, :-1], df_pca['Cluster'])
print(f'Silhouette Score: {silhouette_avg:.4f}')
```

Characteristics: High recency, low frequency, and moderate monetary value.

Characteristics: Medium to high recency, moderate frequency, and moderate monetary value.
Strategy:
Encouragement: Offer incentives to purchase more frequently, such as discounts on the next purchase or bundled offers.

8-Cluster 7: Champions

Characteristics: Very high recency, frequency, and monetary value.
Strategy:
Exclusive Access: Provide early access to new products, special events, and exclusive discounts.
Personalized Experience: Offer highly personalized recommendations and communications.
Feedback and Involvement: Involve them in new product launches and collect their feedback for product development.

# Interpretation and Insights:

Cluster Analysis Summary:

1-Cluster 0: Hibernating

Characteristics: Customers who have not purchased recently, buy infrequently, and have low spending.
Implications: These customers need re-engagement strategies. They might have lost interest or moved to competitors.
Strategies: Reactivation campaigns, personalized offers, and feedback collection to understand reasons for inactivity.

2-Cluster 1: At-Risk

Characteristics: Customers with moderate frequency and spending but have not purchased recently.
Implications: These customers are at risk of becoming inactive. Retention efforts are crucial.
Strategies: Engagement campaigns, personalized discounts, and reminders about previous purchases.

3-Cluster 2: Can't Lose Them

Characteristics: High frequency and spending but have not made recent purchases.
Implications: These are valuable customers with a risk of becoming inactive. High priority for retention.
Strategies: Exclusive benefits, personalized recommendations, and loyalty programs.

4-Cluster 3: About to Slip

Characteristics: Customers who may soon become inactive with moderate frequency and low spending.
Implications: Immediate action is needed to prevent them from slipping away.
Strategies: Special offers, reminder emails, and feedback collection to address concerns.

5-Cluster 4: Loyal Customers

Characteristics: High frequency, recent purchases, and high spending.
Implications: These customers are loyal and crucial for sustained revenue.
Strategies: VIP programs, personalized thank-you notes, and referral incentives.

6-Cluster 5: New Customers

Characteristics: Recently acquired customers with initial moderate spending.
Implications: These customers are still exploring the brand.
Strategies: Welcome campaigns, onboarding series, and first purchase incentives.

7-Cluster 6: Potential Loyalists

Characteristics: Moderate frequency and spending, with recent activity.
Implications: These customers have the potential to become loyal if nurtured properly.
Strategies: Encouragement through targeted offers, engaging content, and loyalty program introduction.

8-Cluster 7: Champions

Characteristics: Very high frequency, spending, and recent activity.
Implications: These are the best customers, providing significant revenue.
Strategies: Exclusive access, personalized experiences, and active feedback involvement.
